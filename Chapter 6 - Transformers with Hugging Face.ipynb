#### Hugging Face

Hugging Face is the company that created the Transformers Python Library. It is popular because it makes it easy for data professionals to access and utilize pretrained LLMs.

They also host the Model Hub, which contains over 1 million pretrained, open-source models (in addition to base models, there are variants, fine-tuned models, experimental models, etc.).

Note: GitHub is the community for uploading and sharing code, and Model Hub is the community for uploading and sharing pretrained models.

Hugging Face workflow:

- Determine the goal:  Different types of tasks use different types of LLMs.
  - Encoder-Only:  Sentiment Analysis, Named Entity Recognition
  - Decoder-Only: Text Generalization
  - Encoder-Decoder: Zero-Shot Classification, Text Summarization
  - Embedding: Document Similarity

- Identify a pretrained model (from Hugging Face's Model Hub): In Model Hub, we can sort models based on popularity, downloads, and so on.


- Specify input data (a single string, a series/column of text,...)

- Apply the pretrained model (input data and view the output(s))

Note: There are additional steps in case we want to optimize the model for our data.

#### Speeding up Transformers Code

- Use GPU instead of CPU (`device = 'mps'`, `device = 'cuda'`)

- Even with the CPU, there are ways to improve the speed.
    - Use a lighter-weight model (a model with fewer parameters)
    - Try fast tokenization (`use_fast = True`)
    - Specify the number of threads we will be using (`torch.set_num_threads(1)`)
    - Disable gradients. Just for predictions, we do not need to remember gradients, as we do not plan to update parameters.

No optimization:

```python
%%time

sentiment_analyzer = pipeline('sentiment-analysis',
                             model = 'distilbert/distilbert-base-uncased-finetuned-sst-2-english',
                             device = -1,
                             truncation = True)

df['Text'].apply(sentiment_analyzer)
```

With optimization:

```python
import torch
```

```python
%%time

sentiment_analyzer = pipeline('sentiment-analysis',
                             model = 'distilbert-base-uncased-finetuned-sst-2-english',
                            use_fast = True,
                             device = -1,
                             truncation = True)

torch.set_num_threads(1)

with torch.no_grad():
    df['Text'].apply(sentiment_analyzer)
```

#### Named Entity Recognition (NER)

NER is used to find and label important information (people, places, organizations, dates,...) in text.

We can perform this using spaCy. However, it works very well with Transformers.

Here, we will be using the default LLM model for NER: BERT (encoder-only).

The implementation is very similar to sentiment analysis (`model = ner`).

Also, we add `aggregation_strategy = 'SIMPLE'`. This tells the model that we want to look at words, not subwords.

In the output, we can also see the likelihood for the entity group (whether it is an organization, location, or so on).

```python
from transformers import pipeline
```

Since we are working on the new analyzer, it is best to enable warnings.

```python
logging.set_verbosity_warning()
```

```python
ner_analyzer = pipeline('ner')
```

```python
### Let's hide a warnings again.

logging.set_verbosity_error()
```

```python
ner_analyzer = pipeline('ner',
                       model = 'dbmdz/bert-large-cased-finetuned-conll03-english',
                       device = -1)
```

```python
text4 = "I ordered an Arnold Palmer at Applebee's in Springfield."
```

```python
ner_analyzer(text4)
```

When you see "#" infront of a word, it is a sub word. We can hide them by adding a argument in the pipeline.

```python
ner_analyzer = pipeline('ner',
                       model = 'dbmdz/bert-large-cased-finetuned-conll03-english',
                       device = -1,
                       aggregation_strategy = 'SIMPLE')
```

```python
ner_analyzer(text4)
```

Here, 'Arnold Palmer' is a miselanious (MISC), 'Applebee' is a organization (ORG), and 'Springfield' is alocation (LOC).

#### Zero-Shot Classification

Now we investigate an application in the transformers library. This is very easy to do with LLMs but not with traditional machine learning.

Zero-Shot classification is used to categorize text without labels quickly.

For this application, the default LLM is BART (encoder-decoder).

Implementation is very similar to sentiment analysis (that is, we first import the pipeline). However, once the classifier is defined, with the text, we will also provide a list of words (labels). Then the model returns scores that classify it into one of those labels (adding up to 1).

The model can do this task as it is a text-to-text model (encoder-decoder model). Thus, it understands input text and labels and picks the labels that are similar and meaningful to the input text.

For the application, we will be downloading a new model from `Model Hub`.

Link: https://huggingface.co/facebook/bart-large-mnlifacebook/bart-large-mnli

```python
classifier = pipeline('zero-shot-classification',
                        model = 'facebook/bart-large-mnli',
                        device = -1)
```

```python
### Texts

text1, text4
```

```python
### Let's classify text1

classifier(text1, ['quote','food & drinks','technology'])
```

Based on the output we can see that this sentence is a "quote".

**Note**: check the "scores". "quote" has the highest score among the labels.

```python
### text4

classifier(text4, ['quote','food & drinks','technology'])
```

Based on the results, text4 is can be categorized (most likely) as "foods & drinks".

#### Text Summarization

The goal here is to make long text bodies more concise.

The default LLM here is BART.

Implementation is again very similar, as we will be using `pipeline`. However, when the `summarizer` is used, a few parameters are given.

This is useful in practice when text summarization is used with sentiment analysis. Text summarization will reduce the text size (truncate text).

For text summarization we will download a trained model from Model Hub.

Link: https://huggingface.co/facebook/bart-large-cnn

```python
summarizer = pipeline('summarization',
            model = 'facebook/bart-large-cnn',
            device = -1)
```

```python
### Sample Text

text5 = """
            The lemon tree produces a pointed oval yellow fruit. Botanically this is a hesperidium, 
            a modified berry with a tough, leathery rind. The rind is divided into an outer colored layer or zest, 
            which is aromatic with essential oils, and an inner layer of white spongy pith. 
            Inside are multiple carpels arranged as radial segments. The seeds develop inside the carpels. 
            The space inside each segment is a locule filled with juice vesicles. 
            Lemons contain many phytochemicals, including polyphenols, terpenes, and tannins.[3] 
            Their juice contains slightly more citric acid than lime juice (about 47 g/L), 
            nearly twice as much as grapefruit juice, and about five times as much as orange juice.[4]
        """
```

```python
### Apply summarizer

summarizer(text5)
```

```python
### We can also specify extra parameters here.

summarizer(text5, min_length = 20, max_length = 50)
```

Note that the last sentence is not complete. This is due to the `max_length`. This can be resolved by changing the `max_length` or changing the other parameters, such as `early stopping`/`length penalty`.

Every time we run the summarizer, we will getthe  same text. We can get different summaries by changing the parameter `do_sample`.
Here, instead of giving a high probability next word, it tries more random words.

```python
summarizer(text5, min_length = 20, max_length = 50, do_sample = True)
```

```python
books['Summary'] = books['Description'].apply(lambda row: summarizer3(row,
                        min_length = 10, max_length = 50,
                        early_stopping = True, length_penalty = 0.80)[0]['summary_text'])
```

#### Text Generation

Text generation is used to write new text based on a prompt.

The default LLM for text generation is GPT (decoder-only).

The implementation is similar as we are using the `pipeline`. However, we need to download a trained model (`gpt2`).

```python
from transformers import pipeline

generator = pipeline('text-generation',
                    model = 'gpt2',
                    device = -1)
```

```python
prompt = "On a hot summer day, I love to drink cold lemonade because"
```

```python
generator(prompt, max_length = 50, num_return_sequences = 1, do_sample = False)

### max_length: maximum number of tokens

### num_return_sequences: number of outputs we need

### do_sample = False: This would give the most likely next word. With "True", they are more random/more creative.
```

Note that senetnces are repeating. This is becase `gpt2` is not a very good model/older model.

A never model may perform much better than this.

```python
### To be more creative, we can try "do_sample = True".

generator(prompt, max_length = 50, num_return_sequences = 1, do_sample = True)
```
