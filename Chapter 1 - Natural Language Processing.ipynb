### Natural Language Processing

In this chapter, we will discuss the basics of natural language processing (NLP), including key concepts, the evolution of NLP, and its applications & Python libraries.

#### Introduction

Definition of NLP: The application of machine learning algorithms to the analysis, understanding, and manipulation of written or spoken examples of human language.

In simple words, using computers to work with text data.

Example: Consider we have customers and their reviews on products. It is easy for humans to read and understand. But computers struggle. That is where NLP comes in.

NLP falls under Artificial Intelligence (AI), which is a field that tries to replicate what humans can do using computers.


#### History of NLP

The fields of NLP has develpoed over the past 70+ years.


1950's:  Translate Russian into English (250 terms and 6 rules).

1960's: The first chatbot is created at MIT. (Early NLP)

1980's:  Statistical machine translation done using probabilistic models instead of rules.

1990's: Penn Treebank was created with 4 million annotated words for NLP researchers. (Traditional NLP)

2010's: Deep learning is paired with NLP.

Closer to 2020: Transformers are introduced and can tackle a variety of NLP tasks. (Modern NLP)


Early NLP: Grammar rules, pattern matching

Traditional NLP: statistical techniques (statistics, probability), machine learning techniques (supervised learning, unsupervised learning)

Modern NLP: Rcurrent-based techniques (RNN, LSTM,...), Transformer-based techniques (LLM, GPT, BERT,...)


First, we focus on traditional NLP, and then we focus on modern NLP, specifically transformer-based techniques.


Note: With modern NLP, what matters most is performance rather than understanding everything (as in traditional NLP).
